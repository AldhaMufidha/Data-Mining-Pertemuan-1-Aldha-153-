```python
#Import Primary Modules:
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

# use the inline backend to generate the plots within the browser
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')  # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

Kode ini digunakan untuk mengimpor library yang diperlukan untuk analisis dan visualisasi data. NumPy digunakan untuk perhitungan numerik, Pandas untuk mengolah data dalam bentuk tabel (dataframe), dan Matplotlib untuk membuat grafik. Selain itu, gaya grafik diatur menggunakan style ggplot agar tampilannya lebih menarik, serta menampilkan versi Matplotlib yang digunakan.

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df_can.head()
```

Kode ini digunakan untuk membaca dataset imigrasi ke Kanada dari sebuah link dan menyimpannya ke dalam dataframe Pandas agar dapat dianalisis.

```python
df_can.set_index('Country', inplace=True)

# Let's view the first five elements and see how the dataframe was changed
df_can.head()
```

Kode ini digunakan untuk menjadikan kolom Country sebagai indeks dataframe sehingga data lebih mudah diakses berdasarkan nama negara. Setelah itu, df_can.head() digunakan untuk menampilkan lima baris pertama data untuk melihat perubahan pada dataframe.

```python
# membuat list tahun
years = list(map(str, range(1980, 2014)))

# menghitung total imigrasi
df_can['Total'] = df_can[years].sum(axis=1)

# mengurutkan dari terbesar
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# ambil 5 negara teratas
df_top5 = df_can.head()

# transpose dataframe
df_top5 = df_top5[years].transpose()

# tampilkan data
df_top5.head()
```

Kode ini digunakan untuk mengurutkan data berdasarkan total imigrasi terbesar, lalu mengambil 5 negara dengan jumlah imigrasi tertinggi. Setelah itu, data diubah posisinya (transpose) agar tahun menjadi baris dan negara menjadi kolom, sehingga lebih mudah digunakan untuk visualisasi.

```python
# mengubah index tahun menjadi tipe integer
df_top5.index = df_top5.index.map(int)

# membuat area chart
df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10))

# memberi judul dan label pada grafik
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# menampilkan grafik
plt.show()
```

Kode tersebut berfungsi untuk memvisualisasikan tren jumlah imigrasi dari lima negara dengan imigrasi terbesar ke Kanada dalam bentuk grafik area berdasarkan tahun, lengkap dengan judul dan label sumbu agar grafik lebih mudah dipahami.

```python
df_top5.index = df_top5.index.map(int)
# option 2: preferred option with more flexibility
ax = df_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

Kode ini digunakan untuk membuat grafik area yang menampilkan tren jumlah imigrasi dari 5 negara dengan imigran terbesar ke Kanada berdasarkan tahun. Baris df_top5.index.map(int) mengubah indeks tahun menjadi angka agar dapat ditampilkan pada sumbu X. Grafik dibuat dengan df_top5.plot(kind='area', stacked=False) sehingga area tiap negara tidak ditumpuk dan tren masing-masing negara terlihat jelas. Judul serta label sumbu ditambahkan menggunakan plt.title(), plt.xlabel(), dan plt.ylabel(), lalu grafik ditampilkan dengan plt.show(). Berbeda dengan kode sebelumnya yang menggunakan objek ax, kode ini menggunakan plt langsung sehingga lebih sederhana.

```python
# Import library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# pilih hanya kolom numerik
X = df_can.select_dtypes(include=['int64','float64'])

# normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA (2 komponen)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

# masukkan ke dataframe
df_pca = pd.DataFrame(pca_result, columns=['PC1','PC2'])

# clustering KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df_pca['cluster'] = kmeans.fit_predict(df_pca)

# style seaborn
sns.set(style="whitegrid")

# pairplot
g = sns.pairplot(
    df_pca,
    hue='cluster',
    palette='Set2',
    height=4,
    corner=True
)

g.fig.suptitle("Pairplot PCA Cluster", y=1.05)

plt.tight_layout()
plt.show()
```

Visualisasi ini menunjukkan pengelompokan negara berdasarkan pola jumlah imigrasi ke Kanada. Dengan menggunakan PCA, data yang awalnya memiliki banyak variabel disederhanakan menjadi dua komponen utama (PC1 dan PC2). Hasil clustering menunjukkan bahwa beberapa negara memiliki pola imigrasi yang mirip sehingga berada dalam kelompok (cluster) yang sama.


```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load dataset
df_can = pd.read_csv(
'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
)

# Ambil hanya data tahun
years = list(map(str, range(1980, 2014)))
df_years = df_can[years]

# Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_years)

# PCA (mengurangi dimensi jadi 2)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

# KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(pca_result)

# Scatter plot cluster
plt.figure(figsize=(10,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c=clusters, cmap='viridis')

plt.title("Cluster Visualization")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.show()
```

Scatter plot ini menunjukkan pengelompokan negara berdasarkan pola imigrasi ke Kanada. Setiap titik mewakili satu negara, dan warna yang berbeda menunjukkan kelompok (cluster) yang memiliki pola data yang mirip.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

df_can['Total'] = df_can.iloc[:, 5:].sum(axis=1)

fig = px.choropleth(
    df_can,
    locations="Country",
    locationmode="country names",
    color="Total",
    hover_name="Country",
    color_continuous_scale="YlOrRd",
    title="Jumlah Imigran ke Kanada Berdasarkan Negara"
)

# hanya 5 negara terbesar
top5 = df_can.sort_values(by='Total', ascending=False).head(5)

fig.add_trace(
    go.Scattergeo(
        locations=top5['Country'],
        locationmode='country names',
        text=top5['Country'],
        mode='markers+text',
        textposition="top center",
        marker=dict(size=6)
    )
)

fig.show()
```

Peta tersebut menampilkan jumlah imigran yang datang ke Kanada dari berbagai negara di dunia. Setiap negara diberi warna berdasarkan total imigrasinya, di mana warna yang lebih gelap menunjukkan jumlah imigran yang lebih besar, sedangkan warna yang lebih terang menunjukkan jumlah yang lebih kecil.
